```python
import numpy as np
import pandas as pd

import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

Loading the mnist dataset

```python
(X_train, y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
len(X_train)
```

```python
len(X_test)
```

Each sample of the size 28*28 pixels

```python
X_train[0].shape
```

Weights represented in the 2d array as below

In this the 0 represnts the black and the 255 means the white
the digit is represented by the white color

all the values lies between the 0 and 255

```python
X_train[0]
```

Here displaying the array of the x_train and the y_train

```python
for i in range(10):
    plt.matshow(X_train[i])
```

```python
y_train[:10]
```

Scalling of the values by dividing by the 255 as values lies between 0 to 255

```python
X_train=X_train/255
X_test=X_test/255
```

Flattening the array into the single 1d array

so we use the reshape funtion to do it

```python
X_train_flattened=X_train.reshape(len(X_train),28*28)
X_test_flattened=X_test.reshape(len(X_test),28*28)

X_train_flattened.shape
```

```python
X_test_flattened.shape
```

Now because of the Flattened the 2d array will look like the 1d array as show below

```python
X_train_flattened[0]
```

Creating the simple
**Neural Network**
with the

input layer = 784 elements &
output layer = 10 elements

for this we use **keras.Sequential()**
sequential means the stack of layer in neural network and it accept every  layer as a element

**dense** means all the input neurons in one layer are connected with every other neuron in the second layer

Optimizer used to train the model efficiently.

loss fucntion

```python
model =keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train_flattened,y_train,epochs=5)
```

Evaluate the accuracy on the test data

```python
model.evaluate(X_test_flattened,y_test)
```

Doing sample prediction on only the first image

```python
plt.matshow(X_test[1])
```

```python
y_predicted=model.predict(X_test_flattened)
y_predicted[1]
```

to look at the maximum score we use the numpy in which there is function called

np.argmax() will find the maximum value only and print the index of that value

```python
np.argmax(y_predicted[1])
```

now the predicted(y_predicted) values these are the whole values so we need to convert it into the concerte class labels

so we use the list comprehension

```python
y_predicted_labels=[np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
y_test[:5]
```

To look how the prediction look like use confusion matrix

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
# prompt: display the confusion matrix using the seaborn

import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

Adding hidden layer

also increase the performance

```python
model =keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train_flattened,y_train,epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
y_predicted=model.predict(X_test_flattened)
y_predicted_labels=[np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

without using the flattened array which can be done by using keras

```python
model =keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train,y_train,epochs=5)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_predicted=model.predict(X_test)
y_predicted_labels=[np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
